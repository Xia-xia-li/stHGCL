```python
import numpy as np
import pandas as pd
import scanpy as sc
import gseapy
import matplotlib.pyplot as plt
import warnings
import os
import urllib.request
import matplotlib as mpl

### ============================ 1. first

# 让 PDF 中的字体可编辑
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['font.family'] = 'Arial'

plt.rcParams.update({'font.size': 12})  # 设置全局字体大小


### ============================== 2. DE

adata = sc.read_h5ad('/root/autodl-tmp/data/mouse_olfactory_bulb/origin/MOB1clust.h5ad')
rna = sc.read_h5ad('/root/autodl-tmp/data/mouse_olfactory_bulb/origin/Slide-seqV2_MoB.h5ad')
# adt = sc.read_h5ad('../../dataset/human_lymph_node/adata_ADT.h5ad')

rna.X=np.nan_to_num(rna.X,nan=np.nan)
sc.pp.highly_variable_genes(rna,n_top_genes=3000,flavor='seurat_v3')
rna=rna[:,rna.var['highly_variable']]
sc.pp.normalize_total(rna)
sc.pp.log1p(rna)
sc.pp.scale(rna)

# adt.X = np.nan_to_num(adt.X,nan=np.nan)
# sc.pp.normalize_total(adt)
# sc.pp.scale(adt)

rna = rna[adata.obs_names, :]
```

```python

rna.obs['y_pred']=adata.obs['kmeans'].values
rna.obs['y_pred']=rna.obs['y_pred'].astype('str')
rna.var_names_make_unique()

# adt.obs['y_pred']=adata.obs['y_pred'].values
# adt.obs['y_pred']=adt.obs['y_pred'].astype('str')
# adt.var_names_make_unique()

# adt.obs['y_pred'] = pd.Categorical(adt.obs['y_pred'], categories=['0','1','2','3','4'], ordered=True)


sc.tl.rank_genes_groups(rna, 'y_pred', method='wilcoxon', key_added = "wilcoxon")
sc.pl.rank_genes_groups(rna, n_genes=25, sharey=False, key = "wilcoxon")
ax = sc.pl.rank_genes_groups_dotplot(
    rna,
    n_genes=3,
    key="wilcoxon",
    groupby="y_pred",
    show=False,
    swap_axes=False,
    cmap='Blues',           # 改成蓝色渐变
    dendrogram=False,       # 去掉连接簇的树状线
    standard_scale='var',    # 按基因标准化颜色（避免颜色偏差）
) # 绘制差异基因气泡图（去掉连接线，蓝色配色）

plt.savefig('./hln_rna.pdf',format='pdf',dpi=600, bbox_inches='tight') ## bbox_inches自动裁剪空白
plt.show()



### ========================== 3.vasiulization gene
# protein='FCGR3A'
# fig=sc.pl.embedding(adt, basis='spatial', color=protein, s=5, return_fig=True)
# ax = fig.axes[0]  # 获取绘图区 axis
# fig.set_size_inches(1.2, 1.2)  # 设置图像尺寸紧凑

# # 移除坐标轴和边框
# ax.set_xticks([])
# ax.set_yticks([])
# ax.set_xlabel('')
# ax.set_ylabel('')
# for spine in ax.spines.values():
#     spine.set_visible(False)

# ax.set_title(protein,fontsize=8)

# plt.savefig('./hln_'+protein+'.pdf',format='pdf',dpi=600,bbox_inches='tight')
# plt.show()





```
